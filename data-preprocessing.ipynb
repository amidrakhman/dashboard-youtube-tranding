**Data Preprocessing**

```python
import pandas as pd
import json
```

```python
data = pd.read_csv('trending.csv')

data.head()
```

```python
#show columns

data.columns
```

```python
#data baru dengan kolom yang diinginkan

data = data[['title','channel_name',  'view', 'like', 'dislike', 'comment', 'trending_time' ]]

data.head()
```

```python
#cek missing value

data.isnull().sum()
```

```python
#isi dengan  nilai 0

data = data.fillna(0)

data.isnull().sum()
```

```python
#membuat kolom trendeing_date berdasarkan kolom trending_date. 

trending = pd.to_datetime(data['trending_time'])
data['trending_date'] = trending.dt.date

#menghapus kolom trending_date
data = data.drop(['trending_time'], axis=1)
data
```

```python
#cek data duplicat

data[['title', 'channel_name']].value_counts().reset_index(name='counts')
```

```python
data_clear = data.drop_duplicates(['title', 'channel_name'], keep='first')
data_clear = data_clear.reset_index(drop=True)

data_clear
```

```python
data_clear[['title', 'channel_name']].value_counts().reset_index(name='counts')
```

```python
#ubah data ke pickle agar ukuran data menjadi kecil

data_clear.to_pickle('clear_data.pickle')
```
